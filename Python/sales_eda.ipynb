```python

# Discount & Promotion Calculations

```

```python
promotion_df.head()
```

```python
promotion_df.info()
```

```python
fact_df = fact_df.merge(
    promotion_df[['PromotionID', 'Price Reduction Type']],
    on='PromotionID',
    how='left'
)
```

```python
fact_df['Discount Percentage'] = 0.0
```

```python
fact_df['Price Reduction Type'] = (
    fact_df['Price Reduction Type']
    .astype(str)
    .str.strip()
    .str.lower()
)
```

```python
# Reset
fact_df['Discount Percentage'] = 0.0
fact_df['Discount Value'] = 0.0

# Extract percentage wherever a number exists
fact_df['Discount Percentage'] = (
    fact_df['Price Reduction Type']
    .str.extract(r'(\d+)')
    .astype(float)
    .fillna(0) / 100
)

# Apply percentage discount ONLY where '%' exists
mask_percent = fact_df['Price Reduction Type'].str.contains('%')

fact_df.loc[mask_percent, 'Discount Value'] = (
    fact_df.loc[mask_percent, 'Total Sales'] *
    fact_df.loc[mask_percent, 'Discount Percentage']
)
```

```python
fact_df['Net Sales'] = fact_df['Total Sales'] - fact_df['Discount Value']
```

```python
# Exploratory Data Analysis (EDA)
```

```python

# What are the Top & Bottom Products by Net Sales?
# Why business cares

# Top products â†’ push marketing, ensure inventory

# Bottom products â†’ fix pricing, promotion, or remove
```

```python
product_sales = (
    fact_df
    .groupby('ProductID', as_index=False)['Net Sales']
    .sum()
)
```

```python
top_5_products = product_sales.sort_values(
    by='Net Sales', ascending=False
).head(5)

top_5_products
```

```python
bottom_5_products = product_sales.sort_values(
    by='Net Sales', ascending=True
).head(5)

bottom_5_products
```

```python
# 
# How do Net Sales change over time?
# Why business cares

# Growth or decline

# Seasonality

# Performance tracking
# 
```

```python
monthly_sales = (
    fact_df
    .groupby(fact_df['Date'].dt.to_period('M'))['Net Sales']
    .sum()
    .reset_index()
)
```

```python
monthly_sales = monthly_sales.sort_values('Date')
monthly_sales.head()
```

```python

# Do discounts hurt or help revenue?
# Why this matters

# Discounts increase volume but may reduce revenue.
```

```python
fact_df['Is Discounted'] = fact_df['Discount Value'] > 0
```

```python
discount_impact = (
    fact_df
    .groupby('Is Discounted')['Net Sales']
    .sum()
    .reset_index()
)

discount_impact
```

```python
#Which promotions are most effective?
```

```python
promotion_performance = (
    fact_df
    .groupby('Price Reduction Type')['Net Sales']
    .sum()
    .reset_index()
    .sort_values('Net Sales', ascending=False)
)

promotion_performance

```
